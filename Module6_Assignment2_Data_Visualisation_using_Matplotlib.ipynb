```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
cars= pd.read_csv('/content/cars.csv')
cars.head()
```

```python
cars.describe()
```

```python
x= cars['carb']
y= cars['gear']
fig= plt.figure(figsize=(15,10))
plt.bar(x,y, linewidth= 2, alpha= 0.7)
plt.xlabel('Number of Carborators')
plt.ylabel('Number of Forward gears')
plt.title('Carbs vs Gears')
plt.grid(True)
plt.show()
```
