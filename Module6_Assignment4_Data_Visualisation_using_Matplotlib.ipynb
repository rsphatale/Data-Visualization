```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
cars= pd.read_csv('/content/cars.csv')
```

```python
plt.pie(cars['cyl'], labels= cars['model'],rotatelabels=True, radius= 2)
plt.pie(cars['cyl'], labels= cars['model'], radius= 3.4)
plt.legend(loc= 'right', bbox_to_anchor= (3,0.5))
plt.show()
```
