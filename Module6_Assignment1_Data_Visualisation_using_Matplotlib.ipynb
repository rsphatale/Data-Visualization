```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cars= pd.read_csv('/content/cars.csv')
```

```python
cars.head()
```

```python
x= cars['model']
y= cars['hp']

fig= plt.figure(figsize= (15,10))
plt.plot(x,y, linewidth= 1, marker= 'o')
plt.xlabel('Models of cars', fontsize= 15)
plt.xticks(rotation= 90)
plt.ylabel('Horse-Power of cars', fontsize= 15)
plt.title('Model Names vs HorsePower', fontsize= 20)
plt.legend(['line'], loc= 'best')
plt.grid(True)
plt.show()
```
