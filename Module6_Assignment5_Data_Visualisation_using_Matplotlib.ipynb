```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
cars= pd.read_csv('/content/cars.csv')
```

```python
x= cars['am']
y= cars['carb']
fig= plt.figure(figsize= (10,5))
plt.fill_between(x, y, color= 'cyan', alpha= 0.6)
plt.plot(x, y, color= 'b')
plt.xlabel('Transmission')
plt.ylabel('Number of Carburator')
plt.title('Transmission vs Number of Carburator')
plt.grid(True)
plt.show()
```
